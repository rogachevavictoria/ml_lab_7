```python
import numpy as np
import pandas as pd
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import datetime as dt
import dateutil as du
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Superstore.csv', delimiter = ';', 
                 parse_dates=['Order Date', 'Ship Date'],encoding='latin1',decimal=",")
df.set_index('Row ID', inplace= True)
df.head()
```

```python
df['State'].unique()[18:21]
```

```python
df_X =df.loc[df['State'] == 'Tennessee']
df_X = df_X.append(df.loc[df['State'] == 'Alabama'])
df_X = df_X.append(df.loc[df['State'] == 'South Carolina'])
df_X.head()
```

```python
df_X =df_X.loc[(df_X['Ship Mode'] == 'First Class') | (df_X['Ship Mode'] == 'Second Class')]
df_X.head()
```

```python
df_X =df_X.loc[df_X['Segment'] == 'Consumer']
df_X.head()
```

```python
X_cols = df.columns[10:10]
X_cols = X_cols.append( df.columns[16:])
X_data = df_X[X_cols]
X_data.head()
```

```python
X_data.describe()
```

```python
X = X_data.values
X_train, X_test = train_test_split(X, test_size=0.2, random_state=27)
print(X_train.shape, X_test.shape)
```

```python
scaler = MinMaxScaler().fit(X_train)
```

```python
X_train_norm = scaler.transform(X_train)
X_train_norm
```

```python
X_train_norm.shape
```

```python
X_test_norm = scaler.transform(X_test)
X_test_norm
```
